if there is 'to' in row_i:
    then split it simply by removing everything to make it something like '99.99129.99'
    Now this string has to be split into 2 separate numbers --- but on what basis?
    Even if I do, I have to then make these into floats to calculate the avg
    return the avg

else there is no 'to' in row_i:
    then as usual, remove everything and make the string as float
    return the float
    


<b>MAKE SURE TO DELETE THE FIRST ROW, WHICH HAS $, FROM THE DF</b>

```python
def clean_price(price):
    
    if 'to' in price:
        price = re.sub(r'[^\d\.]+', '', price)
        parts = price.split('.')
        num1 = float(parts[0] + '.' + str(99))
        num2 = float(parts[1][2:] + '.' + str(99))
        avg = (num1 + num2) / 2
        price = np.round(avg * 96.89, 2)
    
    else:
        price = re.sub(r'[^\d\.]+', '', price)
        price = float(price) * 96.89
        price = np.round(price, 2)
    return price
```
